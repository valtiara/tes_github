# cleansing dan casefolding

```python
def remove_punct(Komen):
      #Menggabungkan karakter pada komentar
      Komen = "".join([char for char in Komen if char not in string.punctuation])
      #Menghilangkan angka
      Komen = re.sub('[0-9]+', '', Komen)
      #Mengganti komen ke huruf kecil  
      Komen = Komen.lower()
      return Komen

# Menggunakan function pada kolom komentar
df['Komentar'] = df['Komentar'].apply(lambda x: remove_punct(x))
df.head(50)
```

# Tokenizing

```python
# Membuat fungsi tokenization untuk memisahkan kalimat menjadi kata
def tokenization(Komen):
  Komen = re.split('\W+', Komen)
  return Komen

#Menerapkan function pada kolom komen
data2['Komentar'] = data2['Komentar'].apply(lambda x: tokenization(x.lower()))
#Menampilkan data
data2.head(50)
```

# Normalization

```python
#TAHAPAN MEMULAI NORMALISASI

normalizad_word = pd.read_csv("D:\skripsi part 2\develop\dataset\kamus_normalisasi.csv", encoding='ISO-8859-1')

#Membuat dictionary
normalizad_word_dict = {}

#Melakukan perulangan dan pengecekan pada tiap baris dataset
for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 
        
#Melakukan normalisasi
def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

#Menggunakan function normalisasi ke kolom komen
data2['Komentar'] = data2['Komentar'].apply(normalized_term)
#Menampilkan 100 data teratas
data2['Komentar'].head(50)
```

```python
#STOPWORD

#Membuat fungsi stopword untuk menghilangkan kata yang tidak perlu /tidak ada makna
def stopwords(Komen):
   reg = re.compile(r"\n")
   return reg.split(Komen)

 # Membuka file contoh stopword
file = open("D:\skripsi part 2\develop\dataset\stopwords.txt","r");
st = file.read()
stopwords = stopwords(st)
stopwords
```

```python
def remove_stopwords(Komen):
   Komen = [word for word in Komen if word not in stopwords]
   return Komen

 #Menerapkan fungsi pada komen 
data2['Komentar'] = data2['Komentar'].apply(lambda x: remove_stopwords(x))
    
 #Menampilkan 50 data teratas
data2.head(50)
```

# Stemming

```python
# import StemmerFactory class
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
st = StemmerFactory()
st_kata = st.create_stemmer()
```

```python
def stemming(text):
    stem_text = [st_kata.stem(word) for word in text]
    return stem_text
# create stemmer

data2['Komentar'] = data2['Komentar'].apply(lambda x: stemming(x))
```
